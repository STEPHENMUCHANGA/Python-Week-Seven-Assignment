```python

# Data Analysis Assignment using the Iris Dataset

# Importing the necessary libraries for the tasks
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Task 1: Load and Explore Iris Dataset

# Load Iris dataset from scikit-learn (earlier known as sklearn) 
try:
    iris = load_iris(as_frame=True)
    df = iris.frame  # this is the dataframe with features + target
    df['species'] = iris.target_names[iris.target]  # Add species column
except Exception as e:
    print("Error loading dataset:", e)

# Display first 10 rows of each species
# Setosa
print("=== First 10 rows: Iris-setosa ===")
display(df[df['species'] == 'setosa'].head(10))

# Save first 10 rows of setosa to an excel file
try:
    df[df['species'] == 'setosa'].head(10).to_excel("First_10_iris_setosa_sample.xlsx", index=False)
    print("First 10 rows of Iris-setosa saved to First_10_iris_setosa_sample.xlsx")
except Exception as e:
    print("Error saving Iris-setosa sample:", e)
    print("Continuing with analysis...")
    # Setosa
print("=== First 10 rows: Iris-setosa ===")
display(df[df['species'] == 'setosa'].head(10))

# Save first 10 rows of setosa to an excel file
try:
    df[df['species'] == 'setosa'].head(10).to_excel("iris_setosa_sample.xlsx", index=False)
    print("First 10 rows of Iris-setosa saved to iris_setosa_sample.xlsx")
except Exception as e:
    print("Error saving Iris-setosa sample:", e)
    print("Continuing with analysis...")

# Versicolor
print("=== First 10 rows: Iris-versicolor ===")
display(df[df['species'] == 'versicolor'].head(10))

# Save first 10 rows of versicolor to an excel file
try:
    df[df['species'] == 'versicolor'].head(10).to_excel("First_10_iris_versicolor_sample.xlsx", index=False)
    print("First 10 rows of Iris-versicolor saved to First_10_iris_versicolor_sample.xlsx")
except Exception as e:
    print("Error saving Iris-versicolor sample:", e)
    print("Continuing with analysis...")

# Virginica
print("=== First 10 rows: Iris-virginica ===")
display(df[df['species'] == 'virginica'].head(10))

# Save first 10 rows of virginica to an excel file
try:
    df[df['species'] == 'virginica'].head(10).to_excel("First_10_iris_virginica_sample.xlsx", index=False)
    print("First 10 rows of Iris-virginica saved to First_10_iris_virginica_sample.xlsx")
except Exception as e:
    print("Error saving Iris-virginica sample:", e)
    print("Continuing with analysis...")

# Check dataset info
print("\nDataset Info:")
print(df.info())

# Check for missing values
print("\nMissing Values per Column:")
print(df.isnull().sum())

# Clean the Iris dataset (no missing values in iris, but were included for demonstration)
df = df.dropna()

# Task 2: Basic Data Analysis for the Iris Dataset

# Descriptive data analysis/statistics (mean, median, mode, std, etc.)
print("\nBasic Statistics:")
print(df.describe())

# Save basic statistics to an excel file
try:
    df.describe().to_excel("iris_basic_statistics.xlsx")
    print("Basic statistics saved to iris_basic_statistics.xlsx")
except Exception as e:
    print("Error saving statistics:", e)
    print("Continuing with analysis...")

# Group by species and calculate mean of numerical columns
species_group = df.groupby("species").mean()
print("\nAverage Measurements per Species:")
print(species_group)

# Save species group statistics to an excel file
try:
    species_group.to_excel("iris_species_group_statistics.xlsx")
    print("Species group statistics saved to iris_species_group_statistics.xlsx")
except Exception as e:
    print("Error saving species group statistics:", e)
    print("Continuing with analysis...")
    
# Observations
print("\nObservations:")
print("- Iris-setosa generally has the smallest petal length/width.")
print("- Iris-virginica tends to have the largest values across most features.")
print("- Iris-versicolor measurements fall in between setosa and virginica.")

# Save observations to a text file
try:
    with open("iris_observations.txt", "w") as f:
        f.write("Observations:\n")
        f.write("- Iris-setosa generally has the smallest petal length/width.\n")
        f.write("- Iris-virginica tends to have the largest values across most features.\n")
        f.write("- Iris-versicolor measurements fall in between setosa and virginica.\n")
    print("Observations saved to iris_observations.txt")
except Exception as e:
    print("Error saving observations:", e)

# Task 3: Data Visualization of the Iris Dataset

# Set the Seaborn style
sns.set(style="whitegrid")

# 1. Line Chart (simulated trend over index for sepal length)
plt.figure(figsize=(8, 5))
plt.plot(df.index, df["sepal length (cm)"], label="Sepal Length")
plt.title("Trend of Sepal Length Across Samples")
plt.xlabel("Sample Index")
plt.ylabel("Sepal Length (cm)")
plt.legend()
plt.show()

# Save line chart to a png file
try:
    plt.savefig("iris_sepal_length_trend.png")
    print("Line chart saved to iris_sepal_length_trend.png")
except Exception as e:
    print("Error saving line chart:", e)
    print("Continuing with analysis...")
    
# 2. Bar Chart (average petal length per species)
plt.figure(figsize=(7, 5))
sns.barplot(x="species", y="petal length (cm)", data=df, palette="muted")
plt.title("Average Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Average Petal Length (cm)")
plt.show()

# Save bar chart to a png file
try:
    plt.savefig("iris_average_petal_length_by_species.png")
    print("Bar chart saved to iris_average_petal_length_by_species.png")
except Exception as e:
    print("Error saving bar chart:", e)
    print("Continuing with analysis...")

# 3. Histogram (distribution of sepal width)
plt.figure(figsize=(7, 5))
plt.hist(df["sepal width (cm)"], bins=15, color="teal", edgecolor="black")
plt.title("Distribution of Sepal Width")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency")
plt.show()

# Save histogram to a png file
try:    
    plt.savefig("iris_sepal_width_distribution.png")
    print("Histogram saved to iris_sepal_width_distribution.png")
except Exception as e:
    print("Error saving histogram:", e)
    print("Continuing with analysis...")

# 4. Scatter Plot (sepal length vs petal length)
plt.figure(figsize=(7, 5))
sns.scatterplot(x="sepal length (cm)", y="petal length (cm)", hue="species", data=df, palette="deep")
plt.title("Sepal Length vs Petal Length by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title="Species")
plt.show()

# Save scatter plot to a png file
try:    
    plt.savefig("iris_sepal_vs_petal_length.png")
    print("Scatter plot saved to iris_sepal_vs_petal_length.png")
except Exception as e:
    print("Error saving scatter plot:", e)
    print("Continuing with analysis...")

# 5. Box Plot (distribution of petal width by species)
plt.figure(figsize=(7, 5))
sns.boxplot(x="species", y="petal width (cm)", data=df, palette="deep")
plt.title("Distribution of Petal Width by Species")
plt.xlabel("Species")
plt.ylabel("Petal Width (cm)")
plt.show()

# Save box plot to a png file
try:    
    plt.savefig("iris_petal_width_distribution_by_species.png")
    print("Box plot saved to iris_petal_width_distribution_by_species.png")
except Exception as e:
    print("Error saving box plot:", e)
    print("Continuing with analysis...")

# 6. Heatmap (correlation matrix of features)
plt.figure(figsize=(8, 6))
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Correlation Matrix of Iris Features")
plt.show()

# Save heatmap to a png file
try:    
    plt.savefig("iris_correlation_matrix.png")
    print("Heatmap saved to iris_correlation_matrix.png")
except Exception as e:
    print("Error saving heatmap:", e)
    print("Continuing with analysis...")
# End of the Data Analysis Assignment using the Iris Dataset
```
